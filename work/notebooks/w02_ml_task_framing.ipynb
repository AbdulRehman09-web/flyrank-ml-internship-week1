```python
y_pred_baseline2 = np.ones(len(y_true))
print(f"\nBaseline (always predict 'declining'):")
print(f"Precision: {precision_score(y_true, y_pred_baseline2):.3f}")
print(f"Recall: {recall_score(y_true, y_pred_baseline2):.3f}")
print(f"F1-score: {f1_score(y_true, y_pred_baseline2):.3f}")
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
declining = df[df['trend_direction'] == 'down']
stable = df[df['trend_direction'] != 'down']
```

```python
print(f"- n = {len(declining)}")
print(f"- CTR range: [{declining['ctr'].min():.2f}, {declining['ctr'].max():.2f}]")
print(f"- Position range: [{declining['avg_position'].min():.1f}, {declining['avg_position'].max():.1f}]")
print(f"- Engagement range: [{declining['engagement_rate'].min():.2f}, {declining['engagement_rate'].max():.2f}]")
```

```python
print(f"- n = {len(stable)}")
print(f"- CTR range: [{stable['ctr'].min():.2f}, {stable['ctr'].max():.2f}]")
print(f"- Position range: [{stable['avg_position'].min():.1f}, {stable['avg_position'].max():.1f}]")
print(f"- Engagement range: [{stable['engagement_rate'].min():.2f}, {stable['engagement_rate'].max():.2f}]")
```

```python
print("Would a rule like 'CTR < 0.5' work?")
rule_positions = df['ctr'] < 0.5
rule_matches_declining = (rule_positions & (df['is_declining'] == 1)).sum()
rule_matches_stable = (rule_positions & (df['is_declining'] == 0)).sum()
```

```python
print(f"Items with CTR < 0.5: {rule_positions.sum()}")
print(f"Declining: {rule_matches_declining}")
print(f"Stable: {rule_matches_stable}")
print(f"False positive rate: {rule_matches_stable / rule_positions.sum():.1%}")
```

```python
print("\nDeclining items by metrics:")
bins = [0, 1, 2, 5, 10, 20, 100]
df['ctr_bin'] = pd.cut(df['ctr'], bins)
print(pd.crosstab(df['trend_direction'], df['ctr_bin'], normalize='columns') * 100)
```

```python
print("\nDeclining rate by content type:")
decline_rate_by_type = df.groupby('content_type')['is_declining'].mean().sort_values()
display(decline_rate_by_type)
```

```python
top_clients = df.groupby('client_id').size().nlargest(10).index
print("\nDeclining rate by client (top 10 clients):")
for client in top_clients:
    client_df = df[df['client_id'] == client]
    decline_rate = client_df['is_declining'].mean()
    print(f"Client {client[:8]}...: {decline_rate:.1%}")
```
